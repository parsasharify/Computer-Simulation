سپس درون این مربع نقاط رندوم را انتخاب کرده و میبینیم که چه تعداد از این نقاط درون دایره قرار میگیرد، با استفاده از این تعداد یک نسبت از مساحت دایره به مربع به دست می‌آوریم، در این جا تنها مورد مجهول ما عدد پی میباشد که میتوان آن را با یک تقسیم ساده پیدا کرد.


```python
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
points_in_circle = 0
total_points = 0
pi = 0
list_of_pi_values = []
x_list = []
real_pi = []
```

```python
i = 0
for i in range(100000):
 
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
 
    distance = x ** 2 + y ** 2
 
    if distance <= 1:
        points_in_circle += 1
 
    total_points += 1
 
    # Estimating value of pi,
    # pi= 4*(no. of points generated inside the
    # circle)/ (no. of points generated inside the square)
    pi = 4 * (points_in_circle / total_points)

    list_of_pi_values.append(pi)
    x_list.append(i)
    real_pi.append(3.141592653589793)
    
 
    print(f"iteration: {i}  pi value: {pi}")
    
    if("{0:.4f}".format(pi) == 3.1415):
      print("{0:.4f}".format(pi))
      break
 
print("Final Estimation of Pi=", pi)
```

```python
plt.plot(x_list, list_of_pi_values)

plt.plot(x_list, real_pi)
  
plt.xlabel('iteration')
plt.ylabel('value')

plt.legend()

plt.show()
```

Part 2 of Q6 for area of oval:

```python
list_of_area_values = []
x_list = []

x_min, x_max = -4, 4
y_min, y_max = -4, 4

num_points = 100000

inside_count = 0

for i in range(num_points):
    x = random.uniform(x_min, x_max)
    y = random.uniform(y_min, y_max)
    if 5*x**2 - 6*x*y + 5*y**2 - 32 <= 0:
        inside_count += 1

    area_fraction = inside_count / (i+1)

    total_area = (x_max - x_min) * (y_max - y_min)

    oval_area = area_fraction * total_area

    x_list.append(i)
    list_of_area_values.append(oval_area)



print("Estimated area of oval:", oval_area)





plt.plot(x_list, list_of_area_values)
  
plt.xlabel('iteration')
plt.ylabel('value')

plt.legend()

plt.show()
```
